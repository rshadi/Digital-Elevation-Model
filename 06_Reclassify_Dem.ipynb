# Reclassify Map

```python
import numpy as np
import matplotlib.pyplot as plt
from osgeo import gdal

# مسیر فایل DEM
path_to_dem_file = "/content/D60624ne.pix"

# خواندن فایل DEM
dem_data = gdal.Open(path_to_dem_file)
dem_array = dem_data.ReadAsArray()

# تعیین threshold برای reclassify
thresholds = [1000, 1200, 1400, 1600, 1800, 2000, 2100, 2200, 2400]

# ایجاد کلاس‌های reclassify
reclassified_dem = np.zeros_like(dem_array)
for i, threshold in enumerate(thresholds):
    reclassified_dem[dem_array >= threshold] = i + 1

# نمایش نقشه reclassified DEM
plt.figure(figsize=(10, 8))
plt.imshow(reclassified_dem, cmap='terrain', extent=(0, dem_array.shape[1], 0, dem_array.shape[0]))
plt.colorbar(label='Elevation Class')
plt.title('Reclassified DEM Map')
plt.xlabel('Column')
plt.ylabel('Row')
plt.grid(True)
plt.show()
```
